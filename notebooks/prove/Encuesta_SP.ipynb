```python
import nltk
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords

# Descargar las stopwords de NLTK si es la primera vez que lo usas
nltk.download('stopwords')

# Cargar las stopwords en espaol 
stop_words = set(stopwords.words('spanish'))  # Se puede cambiar 'spanish' por 'english'

# Agregar stopwords adicionales que consideres innecesarias
custom_stopwords = {'despus', 'patrones', 'dos', 'embargo', 'ninguno', 'despues', 'gustaria', 'misma','mas','soporte', 'usuario', 'inscomexico', 'hola', 'com', 'buenas', 'encuesta', 'ustedes', 'insco', 'gracia', 'da', 'gracias', 'buenos', 'puede', 'favor', 'si', 'claro', 'das'}  # Aade ms palabras aqu si es necesario
stop_words.update(custom_stopwords)

# Leer el archivo Chat_Odoo.txt
with open("survey.txt", "r", encoding="utf-8") as file:
    text_surv = file.read()

# Crear la nube de palabras eliminando las stopwords
wordcloud_surv = WordCloud(stopwords=stop_words, background_color='white', max_words=150).generate(text_surv)

# Mostrar la nube de palabras usando matplotlib
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud_surv, interpolation='bilinear')
plt.axis("off")  # Quitar los ejes
plt.show()
```
